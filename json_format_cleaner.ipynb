```python
import json
import re
import json5
import gradio as gr
from jsonschema import validate, ValidationError
```

```python
def infer_type(value):

    value = value.strip()

    if value.lower() in ["true","yes","on"]:
        return True

    if value.lower() in ["false","no","off"]:
        return False

    if value.lower() in ["null","none","nil"]:
        return None

    try:
        if "." in value:
            return float(value)
        return int(value)
    except:
        return value
```

```python
def clean_json(text):

    data = {}

    lines = text.split("\n")

    for line in lines:

        if ":" not in line:
            continue

        key, value = line.split(":",1)

        key = key.strip().lower().replace(" ","_")
        value = infer_type(value.strip())

        data[key] = value

    return json.dumps(data,indent=4)
```

```python
def remove_nulls(json_text):

    try:
        data = json.loads(json_text)

        cleaned = {k:v for k,v in data.items() if v is not None}

        return json.dumps(cleaned,indent=4)

    except:
        return "Invalid JSON"
```

```python
def validate_json(json_text):

    try:
        json.loads(json_text)
        return "✅ Valid JSON"

    except Exception as e:
        return f"❌ Invalid JSON : {e}"
```

```python
with gr.Blocks() as demo:

    gr.Markdown("# JSON Format Cleaner")

    input_text = gr.Textbox(lines=10,label="Enter Key Value Data")

    output_json = gr.Code(label="Clean JSON")

    clean_btn = gr.Button("Clean JSON")

    clean_btn.click(clean_json,input_text,output_json)

    gr.Markdown("### Validation")

    validate_btn = gr.Button("Validate JSON")

    validation_output = gr.Textbox()

    validate_btn.click(validate_json,output_json,validation_output)

demo.launch()
```
